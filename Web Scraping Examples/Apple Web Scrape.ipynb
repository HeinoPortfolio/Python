#  Apple Web Scrape

This file will open the Yahoo Finance page that contains information on Apple's closing price. IT will retrieve the first 100 listings in the page.  It will display them in a data frame.  This version of the program will use Python 2.7.

````python
import urllib2 as u2
from bs4 import BeautifulSoup
import pandas as pd

def display_Table(stock_table):
    """  Display the stock data
    Args:
        stock_table(list)   list with stock data

    Returns:
        None
    Examples:
        >>>myObject.display_table(stock_table)
    """
    headers = ['Date', 'Close Price']

    pd.set_option('display.max_rows', 500)
    data_frame = pd.DataFrame.from_records(stock_table, columns=headers)

    print data_frame

def main():
    """  Main entry point for the program
    Args:
        None
    Returns:
        None
    Examples:
    >>>main()
    """
    
    url_to_scrape = 'https://finance.yahoo.com/quote/AAPL/history?p=AAPL'
    web_page = u2.urlopen(url_to_scrape)

    soup_output = BeautifulSoup(web_page, "lxml")
    web_page.close()

    table = soup_output.findChildren('table')[0]

    all_rows = table.findChildren('tr')[1:-1]

    date = ''
    close_price = ''

    stock_data = []

    for table_row in all_rows:
        index = 0
        cells = table_row.findChildren('td')
        for cell_content in cells:
            if index == 0:
                date = cell_content.getText()
            elif index == 4:
                close_price = cell_content.getText()
            index += 1
        stock_data.append([date, close_price])

    display_Table(stock_data)

if __name__ == "__main__":
     main()    

 
````

This is the dataframe that contains the listing with the 100 closing prices of Apple.  The string can be changed to look for other stocks. You will need to modify this  ```url_to_scrape = 'https://finance.yahoo.com/quote/AAPL/history?p=AAPL' ```
