# Homework 9 Create Visualizations 

This notebook creates a scatter plot from the Titanic dataset that was given during Homework 4. This dataset has data about the passengers that were onboard the Titanic. This notebook with plot a scatter plot that looks for the the relationship between the variables.

The code belows creates a simple scatter plot of one of the features from the Titanic dataset.  These feature was Age", "Passenger Fare", and "Sex".

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

# Read the fate from the Excel file inot a dataframe.

```python
titanic_df = pd.read_excel('Titanic_Data.xls')
```

# Create the frame that will have the features for the scatter plot.

```python
columns_needed = ["Age", "Passenger Fare", "Sex"]
scatter_plot_df = titanic_df[columns_needed]
scatter_plot_df.dropna(subset=["Age"], inplace=True)
scatter_plot_df.dropna(subset=["Passenger Fare"], inplace=True)
```

# Create the Scatterplot

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x="Age", y="Passenger Fare", hue="Sex", data=scatter_plot_df)
plt.show()
```
