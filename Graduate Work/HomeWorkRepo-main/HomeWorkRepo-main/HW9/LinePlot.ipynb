# Homework 9 Create Visualizations 

This notebook creates a Lineplot from the Titanic dataset that was given during Homework 4. This dataset has data about the passengers that were onboard the Titanic. This notebook with plot a line plot that looks for the the relationship between the variables.

The code belows creates a simple line plot of three features from the Titanic dataset.  These features are Age across the x-axis, the passenger fare across the y-axis, and sex (gender) of the passenger to give the hue of the grap/plot. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Read in the data from the Excel file and show the contents of the frame and get a little info about the set.

```python
# Read the fate from the Excel file inot a dataframe.
titanic_df = pd.read_excel('Titanic_Data.xls')

#Review the data that is in the frame.
titanic_df.head(5)
titanic_df.info()
```

# Create a data frame with only the needed columns for the line plot

```python
columns_needed = ["Age","Passenger Fare", "Sex"]
line_plot_df = titanic_df[columns_needed]
print(line_plot_df)
```

# Create the graph

```python
plt.figure(figsize=(10,10))
sns.lineplot(x="Age", y="Passenger Fare", hue="Sex", style="Sex", data=line_plot_df)
plt.show()
```
