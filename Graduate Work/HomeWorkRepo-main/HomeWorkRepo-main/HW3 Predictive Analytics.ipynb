## Reading in the data file
We need to read in the data from the CSV file. The file name is suv_data.csv. This data will be read into a pandas dataframe to facilitate processing later on.

We should check the contents of the dataframe to make sure that the data has been successfully read from the CSV file and is contained in the frame.

```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
#import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

suv_data_df =  pd.read_csv('suv_data.csv')
```

Check to see what is in the dataframe.

```python
suv_data_df.head(5)
```

Get a little information about what is in the data set.  This code will retrieve the columns, and the type of data that is stored in each of the columns.

```python
suv_data_df.info(verbose=True)
```

To make it easier to use the data we will drop one of the columns that are not needed.  We do not need the User ID column to perfrom the prediction that we need to make. We will also display the dataframe to verify that the column was dropped from the dataframe.


```python
suv_data_df.drop(['User ID'], axis=1, inplace=True)
suv_data_df.head(5)
```

## To encode the Gender column
To use the gender column it should be changed from Male and Female to 0 and 1.  The code will display the modified column to ensure that the column has been changed successfully.

```python
suv_data_df['Gender'].replace(['Male', 'Female'], [1,0], inplace=True)
print(suv_data_df.head(5))
```

## Create the features vector and target variable
We will read the columns that will make of the features the values will come from columns 1,2,0.  These correspond to 

```python
X = suv_data_df.iloc[:, [1,2,0]].values
y = suv_data_df.iloc[:,3].values
```

## Split the data into the training set and test set.
The code to break up the data into sets for prediction.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,
                                                    random_state=0)
```

## Scaling the values using StandardScaler -

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.fit_transform(X_test)
```

## Implement the Logistic Regression Model 
Set the random_state to give consistent results. Allow for more consistency.

```python
log_reg_classifier = LogisticRegression(random_state=42)
log_reg_classifier.fit(X_train, y_train)
y_pred = log_reg_classifier.predict(X_test)
```

## Create a dataframe with predictions and the actual
A dataframe to look at the values as predicted versus what they were.  Show the contents of the dataframe.

```python
results_df = pd.DataFrame({'Test': y_test, 'Predicted': y_pred},
                          columns=['Test', 'Predicted'])

pd.set_option('display.max_rows', None)
results_df
```

## Show a classification report
A report to show the accuracy of the logistic regression model. 

```python
pd.DataFrame(classification_report(y_test, y_pred, output_dict=True)).T
```
